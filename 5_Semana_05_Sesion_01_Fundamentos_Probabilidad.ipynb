### **1. La Probabilidad: El Lenguaje de la Incertidumbre en Ingeniería**

En la práctica profesional de la ingeniería, la incertidumbre es una constante. Ya sea evaluando la resistencia de materiales ante cargas sísmicas (**Ingeniería Civil**), prediciendo la variabilidad climática para la gestión de cultivos (**Ingeniería Agrícola**) o controlando la calidad en líneas de procesamiento masivo (**Ingeniería Agroindustrial**), la probabilidad nos proporciona el lenguaje matemático para cuantificar el riesgo y tomar decisiones informadas. Es el puente que nos permite pasar de la observación de datos pasados a la predicción de eventos futuros.

### **2. Conceptos Fundamentales**

Para formalizar el estudio de fenómenos aleatorios, definimos tres pilares básicos:

*   **Experimento Aleatorio:** Cualquier proceso o acción cuyo resultado no puede predecirse con certeza antes de realizarse, incluso bajo condiciones idénticas.
*   **Espacio Muestral ($\Omega$):** El conjunto de todos los resultados posibles que pueden derivarse de un experimento aleatorio.
*   **Evento ($A$):** Un subconjunto del espacio muestral ($A \subseteq \Omega$), que representa un resultado o conjunto de resultados de interés específico para el análisis.

### **3. Ecuaciones de Probabilidad**

Existen dos enfoques principales para asignar un valor numérico a la posibilidad de que ocurra un evento:

*   **Probabilidad Teórica (Regla de Laplace):** Se aplica cuando todos los resultados del espacio muestral son igualmente probables (equiprobables).
    $$
    P(A) = \frac{\text{Número de resultados favorables a } A}{\text{Número total de resultados posibles en } \Omega}
    $$

*   **Probabilidad Empírica (Frecuencia Relativa):** Se basa en la observación de datos reales tras repetir un experimento un gran número de veces ($n$).
    $$
    P(A) \approx f_r = \frac{n_A}{n}
    $$
    Donde $n_A$ es el número de veces que ocurrió el evento de interés.

## **Celda 3: Código Python (Espacio Muestral, Eventos y Probabilidad Empírica)**

````
import pandas as pd
import numpy as np

# Simulamos el registro histórico de inspección de 500 vigas de concreto (Ing. Civil)
# o lotes de producción (Ing. Agroindustrial).
# Resultados posibles (Espacio Muestral): "Aprobado", "Fallo_Menor", "Fallo_Critico"

np.random.seed(42)
espacio_muestral = ['Aprobado', 'Fallo_Menor', 'Fallo_Critico']
# Asignamos una probabilidad teórica subyacente (80% éxito, 15% fallo menor, 5% fallo crítico)
inspecciones = np.random.choice(espacio_muestral, size=500, p=[0.80, 0.15, 0.05])

df_calidad = pd.DataFrame({'Estado_Inspeccion': inspecciones})

# ==========================================
# CÁLCULO DE PROBABILIDAD EMPÍRICA
# ==========================================
print("--- FRECUENCIAS ABSOLUTAS ---")
frecuencias = df_calidad['Estado_Inspeccion'].value_counts()
print(frecuencias)

print("\n--- PROBABILIDAD EMPÍRICA (Frecuencia Relativa) ---")
prob_empirica = df_calidad['Estado_Inspeccion'].value_counts(normalize=True)
print(prob_empirica.round(4))

# Interpretación Ingenieril:
# El "Espacio Muestral" son los 3 estados posibles.
# Un "Evento" puede ser que una viga tenga un 'Fallo_Critico'.
# La Probabilidad Empírica de 'Fallo_Critico' calculada con nuestros datos es cercana a 0.05 (5%),
# calculada simplemente dividiendo los casos favorables entre el total de inspecciones.
````

### **1. Introducción Práctica: ¿Depende la Calidad del Origen?**

Hasta ahora hemos calculado probabilidades de forma general. Sin embargo, en ingeniería es crucial preguntarse: ¿cambia la probabilidad de un fallo si conocemos el origen del material? Por ejemplo, si una viga de concreto presenta un fallo crítico, ¿es más probable que provenga de un proveedor específico que de otro? Responder a esto nos permite identificar cuellos de botella en la cadena de suministro o fallas sistémicas en procesos de producción agrícola o civil.

### **2. Conceptos Clave**

*   **Probabilidad Condicional:** Es la probabilidad de que ocurra un evento $A$, sabiendo que ya ha ocurrido otro evento $B$. Este concepto nos permite "actualizar" nuestra medida de incertidumbre a medida que obtenemos nueva información o contexto.
*   **Independencia Estadística:** Dos eventos son independientes si la ocurrencia de uno no altera la probabilidad de ocurrencia del otro. En control de calidad, buscamos que los fallos sean eventos independientes y aleatorios; si un fallo depende del proveedor, estamos ante un problema de control de procesos que debe corregirse.

### **3. Ecuaciones Matemáticas**

Para el análisis de estos fenómenos, utilizamos las siguientes definiciones formales:

*   **Probabilidad Condicional de $A$ dado $B$:**
    $$
    P(A|B) = \frac{P(A \cap B)}{P(B)}
    $$

*   **Condición de Independencia:**
    Dos eventos son independientes si y solo si se cumple una de las siguientes condiciones:
    $$
    P(A|B) = P(A) \quad \text{ó} \quad P(A \cap B) = P(A) \cdot P(B)
    $$

---

**Transición al Código:** A continuación, utilizaremos la librería **Pandas** para construir una tabla de contingencia cruzada, lo cual nos permitirá calcular estas probabilidades condicionales a partir de los datos simulados de nuestros proveedores.

## **Celda 4: Código Python (Probabilidad Condicional e Independencia)**

````
# Ahora, supongamos que estas inspecciones provienen de dos proveedores distintos
proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])
df_calidad['Proveedor'] = proveedores

# Tabla de contingencia cruzada
tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)

print("--- TABLA DE CONTINGENCIA ---")
print(tabla_cruzada)

# Cálculo de Probabilidad Condicional
# P(Fallo_Critico | Proveedor_B) = Casos de Fallo Crítico en Prov_B / Total de Prov_B
total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']
fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']

prob_condicional = fallos_prov_b / total_prov_b
print(f"\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}")

# Interpretación: Si la probabilidad condicional es igual a la probabilidad marginal (general),
# decimos que los eventos son INDEPENDIENTES. En la práctica ingenieril, si un proveedor
# tiene una tasa de fallos mucho mayor, la calidad "depende" del proveedor.
````
