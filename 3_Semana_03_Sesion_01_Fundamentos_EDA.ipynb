```markdown
### Explicación del código de detección de valores faltantes

Este bloque de código se encarga de identificar y cuantificar los valores ausentes (NaN) en el DataFrame `datos_ingenieria`, lo cual es un paso crucial en el Análisis Exploratorio de Datos (EDA).

```python
# ==========================================
# PASO 2: VALORES FALTANTES
# ==========================================
print("\n--- 2. DETECCIÓN DE VALORES FALTANTES ---") # Imprime un encabezado para esta sección, indicando el paso actual del EDA.
faltantes = datos_ingenieria.isna().sum() # Utiliza el método `.isna()` para crear un DataFrame booleano (True donde hay NaN, False donde no), y luego `.sum()` para contar los True por columna. El resultado es una Serie con el número de valores faltantes por cada columna.
print(faltantes[faltantes > 0]) # Filtra la Serie `faltantes` para mostrar solo aquellas columnas que tienen al menos un valor faltante (es decir, donde el conteo es mayor que 0). Esto ayuda a visualizar rápidamente dónde se encuentran los problemas.
# Interpretación: Saber cuántos faltan y en qué variables nos dice si debemos # imputar datos o descartar registros antes de calcular estadísticas. # Este comentario explica la importancia de este paso: la detección de valores faltantes es fundamental para decidir cómo proceder con la limpieza y el preprocesamiento de los datos (por ejemplo, si se deben reemplazar los valores faltantes o eliminar las filas/columnas que los contienen) antes de realizar cualquier análisis estadístico posterior.
```


#### **Celda 5: Código Python (Paso 4)**

````
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()

asimetria = skew(resistencia_limpia)
curtosis_val = kurtosis(resistencia_limpia)

print(f"Asimetría de la Resistencia: {asimetria:.2f}")
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
````

```markdown
### Explicación del código de la forma de la distribución (Asimetría y Curtosis)

Este bloque de código calcula la asimetría (`skewness`) y la curtosis (`kurtosis`) de la variable `Resistencia_MPa`, lo que nos ayuda a entender la forma de su distribución. Estos son el Paso 4 del Análisis Exploratorio de Datos (EDA).

```python
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---") # Imprime un encabezado para esta sección, indicando el paso actual del EDA.

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna() # Selecciona la columna 'Resistencia_MPa' del DataFrame `datos_ingenieria` y luego usa `.dropna()` para eliminar cualquier valor faltante (NaN) de esta Serie. Esto es crucial porque la asimetría y la curtosis no se pueden calcular con valores NaN.

asimetria = skew(resistencia_limpia) # Calcula la asimetría de la serie `resistencia_limpia` usando la función `skew` de `scipy.stats`. La asimetría mide la simetría de la distribución.
curtosis_val = kurtosis(resistencia_limpia) # Calcula la curtosis de la serie `resistencia_limpia` usando la función `kurtosis` de `scipy.stats`. La curtosis mide la "puntiagudez" de la distribución y el peso de sus colas.

print(f"Asimetría de la Resistencia: {asimetria:.2f}") # Imprime el valor calculado de la asimetría, formateado a dos decimales.
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}") # Imprime el valor calculado de la curtosis, formateado a dos decimales.

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica. # Explica que un valor de asimetría cercano a cero sugiere que la distribución de los datos es simétrica alrededor de su media.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material). # Explica que una curtosis positiva indica que la distribución tiene colas más "pesadas" y picos más pronunciados que una distribución normal, lo que podría implicar una mayor probabilidad de observar valores extremos (por ejemplo, resistencias de material inusualmente bajas o altas).
```
